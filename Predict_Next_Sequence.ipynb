```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Function to generate a sequence of odd numbers
def generate_odd_sequence(length):
    return [2 * i + 1 for i in range(length)]

# Function to create input-output pairs for training the model
def create_data(sequence, input_length):
    X, y = [], []
    for i in range(len(sequence) - input_length):
        X.append(sequence[i:i+input_length])
        y.append(sequence[i+input_length])
    return np.array(X), np.array(y)

# Generate odd number sequence
sequence_length = 100
odd_sequence = generate_odd_sequence(sequence_length)

# Set input length for the model
input_length = 5

# Create input-output pairs
X, y = create_data(odd_sequence, input_length)

# Build the neural network model
model = Sequential([
    Dense(10, input_shape=(input_length,), activation='relu'),
    Dense(1)
])

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
model.fit(X, y, epochs=100, verbose=2)

# Test the model by predicting the next digit in the sequence
test_input = np.array([odd_sequence[-input_length:]])
predicted_next_digit = model.predict(test_input)
print(f"Predicted next digit: {int(predicted_next_digit[0, 0])}")
```
